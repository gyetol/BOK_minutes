# 워드 클라우드

```python
import pandas as pd
import numpy as np
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import matplotlib
from wordcloud import WordCloud, STOPWORDS
```

```python
df_Pos = pd.read_csv('pos.csv')
```

```python
p_text =''
for i in range(0,100):
    p_text += np.int(df_Pos['up_count'][i]) * (df_Pos['ngram'][i].replace("/","").replace("NNG","").replace("VV","").replace("VA","").replace("XSY","").replace(";","").replace("MAG","")+" ")
import random
p_text_list = p_text.split()
random.shuffle(p_text_list)
#gtext_list[:10]    gtext 목록 확인 
wordcloud = WordCloud(font_path='font/NanumGothic.ttf',
                      width= 500,
                      height = 500,
                      background_color='black',
                      colormap='Reds').generate(" ".join(p_text_list))
plt.figure(figsize=(5,5)) #이미지 사이즈 지정
plt.imshow(wordcloud, interpolation='lanczos') #이미지의 부드럽기 정도
plt.axis('off') #x y 축 숫자 제거
plt.show()
```

```python
df_Neg = pd.read_csv('Neg.csv')
```

```python
n_text =''
for i in range(0,200):
    n_text += np.int(df_Neg['up_count'][i]) * (df_Neg['ngram'][i].replace("/","").replace("NNG","").replace("VV","").replace("VA","").replace("XSY","").replace(";","").replace("MAG","")+" ")
import random
n_text_list = n_text.split()
random.shuffle(n_text_list)
#gtext_list[:10]    gtext 목록 확인 
wordcloud = WordCloud(font_path='font/NanumGothic.ttf',
                      width= 500,
                      height = 500,
                      background_color='white',
                      colormap='Blues').generate(" ".join(n_text_list))
plt.figure(figsize=(5,5)) #이미지 사이즈 지정
plt.imshow(wordcloud, interpolation='lanczos') #이미지의 부드럽기 정도
plt.axis('off') #x y 축 숫자 제거
plt.show()
```
