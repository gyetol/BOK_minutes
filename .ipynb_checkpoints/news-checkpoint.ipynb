```python
# -*- coding: utf-8 -*-
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import datetime
error_date=[]
def get_news(n_url):
    header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'}
    html = requests.get(n_url, headers=header)
    bsoup = BeautifulSoup(html.content, 'html.parser')
    # 기사 제목
    title = bsoup.select('h3#articleTitle')[0].text
    # 날짜 
    pdate = bsoup.select('.t11')[0].get_text()[:10]
    # 기사 본문
    _text = bsoup.select('#articleBodyContents')[0].get_text().replace('\n', " ")
    btext = _text.replace("// flash 오류를 우회하기 위한 함수 추가 function _flash_removeCallback() {}", "") #제거
    return title, pdate, btext
def naver_news_crawling(query, start_date, end_date, s_from, e_to, page):
    title_list=[]
    date_list=[]
    contents_list=[]
    while page<50:
        #print(page)
        url="https://search.naver.com/search.naver?&where=news&query="+query+"&sm=tab_pge&sort=2&photo=0&field=0&reporter_article=&pd=3&ds="+start_date+"&de="+ end_date +"&docid=&nso=so:da,p:from"+s_from+"to"+e_to+",a:all&mynews=1&start="+str(page)+"&refresh_start=0"
        #print(url)
        #header 추가 
        header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36',
                 'Cookie' : 'NNB=ZL5DETT7USSF6; ASID=3bba7602000001759c8026f20000005f; nso_open=1; NRTK=ag#20s_gr#3_ma#-2_si#0_en#1_sp#-2; _fbp=fb.1.1606212706103.1193446282; _ga_7VKFYR6RV1=GS1.1.1607688578.19.0.1607688578.60; _ga=GA1.1.1134179991.1604669604; NDARK=N; nx_ssl=2; BMR=; _naver_usersession_=Eav7SrhgCFn75H0D2aVQGw==; page_uid=U/RH3dprvmZssOdKZzossssssDw-349971; news_office_checked=1001,1018,2227; news_my_status=1'}
        req = requests.get(url,headers=header)
        #print(url)
        cont = req.content
        soup = BeautifulSoup(cont, 'html.parser')
        for urls in soup.select("a.info"):
            try :
               #print(urls["href"]) #모든링크
                if urls["href"].startswith("https://news.naver.com"):
                    #print(urls["href"]) #네이버뉴스링크만 뽑기
                    title_text, date_text, contents_text = get_news(urls["href"])
                    title_list.append(title_text)
                    date_list.append(date_text)
                    contents_list.append(contents_text)
            except Exception as e:
                print(e) 
                continue
        page += 10
    result={}   
    result= {'date' : date_list, 'title':title_list,'contents': contents_list}
    df=[]
    df=pd.DataFrame(result)
    return df
#초기값 설정
page = 1
query = "금리" 
date = datetime.date(2010,1,1) #시작날짜 
start_date=date.strftime('%Y.%m.%d')
end_date=start_date
s_from = start_date.replace(".","")
e_to = start_date.replace(".","")
standard_date=start_date
#------------------------------------
#---------------메인으로 돌아가는 부분----------------------
while(standard_date!="2012.12.31"): #끝날짜 2018.01.01
    try: 
        print("---------------"+standard_date+"날짜의 크롤링을 시작합니다---------------")
        new_df = naver_news_crawling(query, start_date, end_date, s_from, e_to, page)
        new_df=new_df.drop_duplicates()
        new_df.to_csv(r"C:\Users\nam28\kdigital\Project\news\2010.csv",encoding='utf-8',index=False, header=False, mode='a')
        date += datetime.timedelta(days=1)
        start_date=date.strftime('%Y.%m.%d')
        end_date=start_date
        s_from = start_date.replace(".","")
        e_to = start_date.replace(".","")
        standard_date=start_date
        print("성공")
    except Exception as e:
        error_date.append(standard_date)
        print(e)
        print("실패")
        continue
print("---------------크롤링을 종료합니다---------------")
```
