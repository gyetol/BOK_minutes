#####  긍정 부정 사전  

```python
gdic =  pd.read_csv('긍정사전1.csv', header = 0)
bdic =  pd.read_csv('부정사전1.csv', header = 0)
```

##### 긍정 사전 클라우드

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

cnt = 0
gtext =''

for i in range(0,1):
    gtext += np.int(gdic['up_count'][i]) * (gdic['ngram'][i].replace("/","").replace("NNG","").replace("VV","").replace("VA","").replace("XSY","").replace(";","").replace("MAG","")+" ")
for i in range(3,300):
    gtext += np.int(gdic['up_count'][i]) * (gdic['ngram'][i].replace("/","").replace("NNG","").replace("VV","").replace("VA","").replace("XSY","").replace(";","").replace("MAG","")+" ")

    #np.int 이유 : float 타입은 문자 곱셈이 안됨 
    #for문을 2번 돌린 이유 : 2번항목을 피하기 위함 -> 추가 코드 수정 필요 
    
import random 
gtext_list = gtext.split()
random.shuffle(gtext_list)

    # 워드클라우드 문자 중복으로 나와서 글자를 랜덤으로 셔틀 돌림 -> 그후 중복이 안됨 

wordcloud = WordCloud(font_path='font/NanumGothic.ttf',
                      width= 500,
                      height = 500,
                      background_color='white',
                      colormap='Reds').generate(" ".join(gtext_list))

plt.figure(figsize=(5,5)) #이미지 사이즈 지정
plt.imshow(wordcloud, interpolation='lanczos') #이미지의 부드럽기 정도
plt.axis('off') #x y 축 숫자 제거
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


btext =''
for i in range(0,300):
    btext += np.int(bdic['up_count'][i]) * (bdic['ngram'][i].replace("/","").replace("NNG","").replace("VV","").replace("VA","").replace("XSY","").replace(";","").replace("MAG","")+" ")

import random
btext_list = btext.split()
random.shuffle(btext_list)
#gtext_list[:10]    gtext 목록 확인 

wordcloud = WordCloud(font_path='font/NanumGothic.ttf',
                      width= 500,
                      height = 500,
                      background_color='white',
                      colormap='Blues').generate(" ".join(btext_list))

plt.figure(figsize=(5,5)) #이미지 사이즈 지정
plt.imshow(wordcloud, interpolation='lanczos') #이미지의 부드럽기 정도
plt.axis('off') #x y 축 숫자 제거
plt.show()
```
